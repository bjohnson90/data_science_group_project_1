```python
%matplotlib inline
import matplotlib.pyplot as plt
xp_vals = [206826,45450,487]
exp_labels = ["2","3","4"]
plt.pie(xp_vals,labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(xp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=2,explode=[0, .2,.4])
plt.show()
```
