```python
%matplotlib inline
import matplotlib.pyplot as plt
xp_vals = [252889,2009,2140,25671,11908,148,1559,221,3237,557,187,168,274,1,20,289,3,4,2,16]
exp_labels = ["201","222","229","241","245","206","244","203", "246", "202", "236","247","343", "200","341","406","239","248","336","339"]
plt.pie(xp_vals,labels=exp_labels)
plt.show()
```

```python
plt.pie(xp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

```python
plt.axis("equal")
plt.pie(xp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=3)
plt.show()
```

```python
plt.axis("equal")
plt.pie(xp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=3,explode=[0.2, 1.2, 1, 0,0,1.2,1.4,1.6,1.8,2,2.2,2.4,2.6,2.8,3,3.2,3.4,3.6,3.8,4])
plt.show()
```
